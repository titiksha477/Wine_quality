##Importing libraries


```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error
import numpy as np
```

##Uploading data


```python
df = pd.read_csv("/content/winequality-red.csv")
```

```python
feature = ['fixed acidity', 'volatile acidity', 'citric acid',
       'chlorides',  'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
```

##Selecting features and Output


```python
x = df[feature]
```

```python
x
```

```python
y = df['quality']
```

##splitting data for training and testing


```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.24, random_state = 0)
```

## Modelling

```python
Lin = LinearRegression()
```

```python
Lin.fit(x_train,y_train)
```

```python
y_pred =Lin.predict(x_test)
```

```python
predited_y= np.round(y_pred)
```

```python
mean_absolute_error(y_test, predited_y)
```

```python
mean_squared_error(y_test, predited_y)
```

```python
accuracy_score(y_test,predited_y)
```
